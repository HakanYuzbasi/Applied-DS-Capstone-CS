### In the following call we will create a function which will extract the category from the FourSquare vanues. Later we will apply FourSquare API in all the neighbourhoods on London.

```python
# Gets the name of the category

def get_category_type(row):
    categories_list = row['Category']  
    
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
explore_df_list = []

for i, nhood_name in enumerate(lon_data['Neighbourhood']):  
    
    try :
        #Get neighbourhood data
        nhood_name = lon_data.loc[i, 'Neighbourhood']
        nhood_lat = lon_data.loc[i, 'Latitude']
        nhood_lng = lon_data.loc[i, 'Longitude']

        radius = 1000 
        LIMIT = 30 

        url = 'https://api.foursquare.com/v2/venues/explore?client_id={} \
        &client_secret={}&ll={},{}&v={}&radius={}&limit={}'\
        .format(CLIENT_ID, CLIENT_SECRET, nhood_lat, nhood_lng, VERSION, radius, LIMIT)

        results = json.loads(requests.get(url).text)
        results = results['response']['groups'][0]['items']

        nearby = json_normalize(results) # Flatten JSON

        # Filter the columns
        filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
        nearby = nearby.loc[:, filtered_columns]

        # Rename the columns
        columns = ['Name', 'Category', 'Latitude', 'Longitude']
        nearby.columns = columns

        # Get the categories
        nearby['Category'] = nearby.apply(get_category_type, axis=1)    

        # Get the required data
        for i, name in enumerate(nearby['Name']):
            s_list = nearby.loc[i, :].values.tolist()  # Converts the numpy array to a python list
            f_list = [nhood_name, nhood_lat, nhood_lng] + s_list
            explore_df_list.append(f_list)
    
    except Exception as e:
        pass
```

### Now we will create a dataframe for performing clustering.

```python
explore_lon = pd.DataFrame([item for item in explore_df_list])
explore_lon.columns = ['Neighbourhood', 'Neighbourhood Latitude', 'Neighbourhood Longitude', 'Venue Name', 'Venue Category', 'Venue Latitude', 'Venue Longitude']
explore_lon.head()
```

### Now we will perform one-hot encoding to analyze neighbourhoods of London.

```python
lon_1hot = pd.get_dummies(explore_lon[['Venue Category']], prefix="", prefix_sep="")


# Add neighbourhood column back to dataframe
lon_1hot['Neighbourhood'] = explore_lon['Neighbourhood'] 


# Move neighbourhood column to the first column
fixed_columns = [lon_1hot.columns[-1]] + lon_1hot.columns[:-1].values.tolist()
lon_1hot = lon_1hot[fixed_columns]

lon_1hot.head()
```

### In the following step we will agregate the venues by neighbourhoods and later we will create a function to return most common venues.

```python
lon_grouped = lon_1hot.groupby('Neighbourhood').mean().reset_index()
lon_grouped.head()
```

```python
def top_common_venues(row, top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:top_venues]
```

### Now we will create a new dataframe to get the top 10 venues.


```python
top_venues = 10
indicators = ['st', 'nd', 'rd']

# Create columns according to number of top venues
columns = ['Neighbourhood']
for ind in np.arange(top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# Create a new dataframe
neighbourhoods_venues_sorted = pd.DataFrame(columns=columns)
neighbourhoods_venues_sorted['Neighbourhood'] = lon_grouped['Neighbourhood']

for ind in np.arange(lon_grouped.shape[0]):
    neighbourhoods_venues_sorted.iloc[ind, 1:] = top_common_venues(lon_grouped.iloc[ind, :], top_venues)

neighbourhoods_venues_sorted.head()
```

### Now we will apply K-means clustering on the dataframe.


```python
lon_grouped_clustering = lon_grouped.drop('Neighbourhood', 1)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

def plot(x, y, xlabel, ylabel):
    plt.figure(figsize=(20,10))
    plt.plot(np.arange(2, x), y, 'o-')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(np.arange(2, x))
    plt.show()
```

```python
max_range = 15 #Max range 15 (number of clusters)

from sklearn.metrics import silhouette_samples, silhouette_score


indices = []
scores = []

for lon_clusters in range(2, max_range) :
    
    # Run k-means clustering
    lon_gc = lon_grouped_clustering
    kmeans = KMeans(n_clusters = lon_clusters, init = 'k-means++', random_state = 0).fit_predict(lon_gc)
    
    # Gets the score for the clustering operation performed
    score = silhouette_score(lon_gc, kmeans)
    
    # Appending the index and score to the respective lists
    indices.append(lon_clusters)
    scores.append(score)
```

```python
plot(max_range, scores, "No. of clusters", "Silhouette Score")
```

### From the graph the optimal number is found to be 8 using the Elbow Method.


```python
opt_value = 8
```

### Now that we have calculated out optimum value of clusters, we can proceed with K-Means clustering.

```python
lon_clusters = opt_value

# Run k-means clustering
lon_gc = lon_grouped_clustering
kmeans = KMeans(n_clusters = lon_clusters, init = 'k-means++', random_state = 0).fit(lon_gc)
```

```python
# Add clustering labels
neighbourhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
```

### Now we will crate a final version of dataframe containing all the necessary data.


```python
lon_final = lon_data
lon_final = lon_final.join(neighbourhoods_venues_sorted.set_index('Neighbourhood'), on='Neighbourhood')
lon_final.dropna(inplace = True)
lon_final['Cluster Labels'] = lon_final['Cluster Labels'].astype(int)
lon_final.head()
```
